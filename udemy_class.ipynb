```
import random
import time

class Kumanda():

  def __init__(self, tv_durum = "kapalı", tv_ses = 0, kanal_listesi = ["trt"], kanal = "trt"):
    print("kumanda oluşturuluyor...")

    self.tv_durum = tv_durum
    self.tv_ses = tv_ses
    self.kanal_listesi = kanal_listesi
    self.kanal = kanal 
  
  def open_tv(self):
    if (self.tv_durum == "kapalı"):
      print("televizyon açılıyor...")
      self.tv_durum = "açık"
    else:
      print("tv zaten açık")

  def close_tv(self):
    if (self.tv_durum == "açık"):
      print("televizyon kapanıyor...")
      self.tv_durum = "kapalı"
    else:
      print("tv zaten kapalı")
  
  def tv_sound(self):

    while True:
      krk = input("sesi aç: 'a'\nsesi kıs: 'k'\nsesi sıfırla: 'm'\n çıkış: 'q'\nişlem: ")

      if (krk == "k"):
        if (self.tv_ses != 0):
          self.tv_ses -= 1
          print("ses:", self.tv_ses)
      
      elif (krk == "a"):
        if (self.tv_ses != 32):
          self.tv_ses += 1
          print("ses:", self.tv_ses)

      elif (krk == "q"):
        break
      
      else: 
        print("geçersiz işlem")
  

  def add_kanal(self, kanal):

    print("kanal ekleniyor...")
    time.sleep(1)
    print("kanal eklendi", kanal)
    self.kanal_listesi.append(kanal)
    

  def __str__(self):
    return "TV_DURUM {},\nSES {},\nLİSTE {}".format(self.tv_durum,self.tv_ses,self.kanal_listesi)

kumanda = Kumanda()

print("""
TV UYGULAMASI

1. TV AÇ
2. TV KAPAT
3. TV SES
4. KANAL EKLE
5. TV BİLGİLERİ

çıkmak için 'ç' tuşuna basın.
""")

while True:

  işlem = input("işlem no giriniz: ")

  if (işlem == "ç"):
    print("program saonlandırıldı")
    break

  if (işlem == "1"):
     kumanda.open_tv()

  elif (işlem == "2"):
     kumanda.close_tv()

  elif (işlem == "3"):
     kumanda.tv_sound()

  elif (işlem == "4"):
     kanal_isimleri = input("kanal isimlerini ',' ile ayırarak girin")
     kanal_listesi = kanal_isimleri.split(",")

     for i in kanal_listesi:
       kumanda.add_kanal(i)
     print("kanal listesi güncellendi") 

  elif (işlem == "5"):
     print(kumanda)
  
  else:
    print("geçersiz işlem")


```

```
import random
import time

class Computer():

  def __init__(self, marka = "Dell", işlemci = "i5", ram = 8):
    print("bilgisayar açılıyor...")
    time.sleep(1)
    print("bilgisayar açıldı, hoşgeldiniz")

    self.marka = marka
    self.işlemci = işlemci
    self.ram = ram
  
  def işlemci_değiştir(self, işlemci):
    
    while True:

      a = input("i7 veya i9 işlemcilerden birini seçiniz.\nçıkmak için 'q' tuşuna basın.\nişlemci seç:")

      if (a == "i7"):
        self.işlemci = "i7"
        break
      elif (a == "i9"):
        self.işlemci = "i9"
        break
      elif (a == "q"):
        print("işlemci değişimi durduruldu")
        break
      else:
        print("hatalı giriş")
    
    print(self.işlemci)
  
  def ram_artı(self, ram):

    while True:
      a = int(input("4 veya 8 gb ramlerden birini seçiniz.\nçıkmak için '4444' tuşuna basın.\nram seç:"))
      
      if (a == 4444):
        print("ram ekleme sonlandırıldı")
        print(ram)
        break
      if (self.ram < 32):
        self.ram += a
        print("ram:", self.ram)
      else:
        print("daha fazla ram eklenemez")

    print(self.ram)

  def __str__(self):
    return "MARKA {},\nİŞLEMCİ {},\nRAM {}".format(self.marka,self.işlemci,self.ram)

  

computer = Computer()

print("""
*************
Bilgisayar

1.Bilgisayarı Aç
2.İşlemci Değiştir
3.RAM Arttır

Çıkmak için 'Q' tuşuna basınız

*************
""")

while True:
  işlem = input("bir işlem girin: ")

  if (işlem == "Q"):
    print("işlemler duraklatıldı")
    break
  elif (işlem == "1"):
    print(computer)
  elif (işlem == "2"):
    computer.işlemci_değiştir("işlemci")
  elif (işlem == "3"):
    computer.ram_artı("ram")
  else:
    print("hatalı işlem")
```

```
class Animal():

  def __init__(self, ayak, kuyruk, göz):
    self.ayak = ayak
    self.kuyruk = kuyruk
    self.göz = göz
  
  def bilgileri_göster(self):
    print(f"ayak: {self.ayak}, kuyruk: {self.kuyruk}, göz: {self.göz}")
  
  def evrim(self, yüzgeç):
    self.ayak = yüzgeç

class Horse(Animal):
  pass

horse = Horse(4, 1, 2)

print(horse.bilgileri_göster())

```
